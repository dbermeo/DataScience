```python
import nltk
import string
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
```

```python

from google.colab import files
file = files.upload()
```

```python
texto=open('eje.txt').read()
```

```python
frases= sent_tokenize(texto)
print(frases)
```

```python
palabras = word_tokenize(texto)
print (palabras)
```

```python
fdist=FreqDist(palabras)
fdist.most_common(20)
```

```python
stop_words=stopwords.words("spanish")
stop_words
```

```python
palabras=word_tokenize(texto.lower())
palabras2=[x for x in palabras if x not in stop_words]
palabras2
```

```python
palabras2=[x for x in palabras if x not in stop_words + list(string.punctuation)]
palabras2
```

```python
FreqDist(palabras2).most_common(20)
```

```python
palabras2
```

```python
from nltk.stem import SnowballStemmer
stemer=SnowballStemmer("spanish")
```

```python
raiz=[]
for palabra in palabras2:
    raiz.append(stemer.stem(palabra))

```

```python
FreqDist(raiz).most_common(20)
```

```python
FreqDist(palabras2).most_common(20)
```

```python
#!pip install googletrans
from googletrans import Translator
```

```python
translator = Translator()
```

```python
print(translator.detect("hola como estas"))
```

```python
print(translator.translate("hola como estas"))
```

```python
peliculas = translator.translate(["el senor de los anillo", "armagedon", "El retorno del rey"])
```

```python
for pelicula in peliculas:
  print(pelicula.origin, "==>",pelicula.text)
```
