```python
consumidores=[("A","Movil"),("B","Movil"),("A","Portatil"),("A","Tablet"),
              ("B","Tablet"),("C","Portatil"),("D","Smartwatch"),("E","Console")]
con_labels=['Consumidores','Productos']
con_df=pd.DataFrame.from_records(consumidores,columns=con_labels)
```

```python
con_df
```

```python
productores=[("a","Movil"),("a","Smartwatch"),("a","Tablet"),("b","Portatil"),
              ("c","Sobremesa"),("c","Portatil")]
prod_labels=["Productores","Productos"]
prod_df=pd.DataFrame.from_records(productores,columns=prod_labels)
```

```python
prod_df
```

```python
pd.merge(con_df,prod_df,on='Productos',how='outer')
```

```python
pd.merge(con_df,prod_df,on='Productos',how='inner')
```

```python
pd.merge(con_df,prod_df,on='Productos',how='right')
```

```python
pd.merge(con_df,prod_df,on='Productos',how='left')
```
