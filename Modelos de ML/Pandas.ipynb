```python
#Test de chi cuadrado
```

```python
np.random.seed(0)
df1=df1[df1["Origin"].isin(['HOU',"ATL","IND"])]
df1=df1.sample(frac=1)
df1=df1[0:10000]
```

```python
df1
```

```python
df1['BigDelay']=df1["ArrDelay"]>30
observados=pd.crosstab(index=df1["BigDelay"],columns=df1["Origin"],margins=True)
observados
```

```python
from scipy.stats import chi2_contingency
```

```python
test=chi2_contingency(observados)
```

```python
test
```

```python
#0.5916164055824548  suma de las diferencias al cuadrado
#la tabla es de valores esperados
```

```python
esperados=pd.DataFrame(test[3])
esperados
```

```python
esperados_rel=round(esperados.apply(lambda r:r/len(df1)*100,axis=1),2)
```

```python
esperados_rel
```

```python
observados_rel=round(observados.apply(lambda r:r/len(df1)*100,axis=1),2)
```

```python
observados_rel
```

```python
test[1]
```

```python
#si p-valor < 0.05 hay diferencias significativas. Hay relacion entre variables
#si p-valor > 0.05 no hay diferncias significativa No hay relacion entre variables
```

```python
df=pd.read_csv("D:/Windows/User/Documentos/ciencia de datos/Recursos/2008.csv",nrows=100000)
```

```python
x= df['ArrDelay'].dropna()
```

```python
Q1=np.percentile(x,25)#separa el 25 duperior de 75% inferrior
Q3=np.percentile(x,75)#separa el 75 duperior de 25% inferrior
rangointer=Q3-Q1 #rango intercuartilico es la diferencia entre el primer cuartil y el tercero
```

```python
umbralsuperior=Q3+1.5*rangointer
umbralinferior=Q1-1.5*rangointer
```
