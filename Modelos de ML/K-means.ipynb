```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:/Windows/User/Documentos/ciencia de datos/Recursos/2008.csv",nrows=1e5)
newdf=df[['AirTime','DepDelay']].dropna()
```

```python
newdf=df[['AirTime','Distance','TaxiOut','ArrDelay','DepDelay']].dropna()
```

```python
kmeans= KMeans(n_clusters=4,random_state=0, n_jobs=-1).fit(newdf)
kmeans.labels_
```

```python
np.unique(kmeans.labels_,return_counts=True)
```

```python
plt.scatter(newdf.AirTime,newdf.DepDelay,c=kmeans.labels_)
```

```python
kmeans.cluster_centers_
```

```python
alldf=pd.read_csv("D:/Windows/User/Documentos/ciencia de datos/Recursos/2008.csv",nrows=1e6)[['AirTime','Distance','TaxiOut','ArrDelay','DepDelay']].dropna()
```

```python
kmeans.predict(alldf)[0:50]
```
