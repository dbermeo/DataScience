```python
df3 = df1.dropna(subset = ['cumulative_total_cases','daily_new_cases','active_cases','cumulative_total_deaths','daily_new_deaths'])
df3 = df3.sample(frac=1).head(5000)
X=df3[['cumulative_total_cases','daily_new_cases','active_cases','cumulative_total_deaths']]
Y=df3['daily_new_deaths']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.2, random_state=1)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train= scaler.transform(X_train)
X_test= scaler.transform(X_test)
```

```python
clf= MLPRegressor(solver= 'lbfgs', alpha=1e-5, hidden_layer_sizes=(5,))
model = clf.fit(X_train,Y_train)
predictions=model.predict(X_test)
print("R cuadrado",r2_score(Y_test,predictions))
```
